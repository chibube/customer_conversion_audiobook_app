```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data = pd.read_csv('Audiobooks_data_title.csv')
```

```python
data.head()
```

```python
 data['Targets']=data['Unnamed: 11'] 
```

```python
data.head()
```

```python
#dropping Customer_ID column

data2 = data.drop(columns=['Cusomer ID','Unnamed: 11'], axis = 1)
data2
```

```python
data2.describe()
```

```python
data2.info()
```

```python
sns.pairplot(data2, hue='Targets')
```

from the scatter plots we can observe that average book length and price paid overall or average have a high impact on conversion.

```python
data2.corr()
```

```python
#plotting correlation heatmap
corr = data2.corr()
# Plot correlation
plt.figure(figsize=(20,18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot = True, annot_kws={'size':10})
# Axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

from the correlation matrix we can see that average book length and average price paid have a strong linear correlation with conversion. Also, the difference between the last visit and purchase date has a high correlation
